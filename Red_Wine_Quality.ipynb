```python
import pandas as pd
df = pd.read_csv('winequality-red.csv', sep=';')
df.head()
```

```python
X = df.drop('quality', axis=1)
y = 'quality'
```

```python
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
scaled_df.head()
```

```python
scaled_df.var()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_df, df[y], test_size=0.1, random_state=34)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
classifiers = [('Random Forest', RandomForestClassifier()), ('Decisio Tree Classifier', DecisionTreeClassifier()), ('XGB Classifier', xgb.XGBClassifier()), ('Naive Bayes', GaussianNB())]
```

```python
from sklearn.metrics import accuracy_score
```

```python
for clf_name, clf in classifiers:
    clf.fit(X_train, y_train)
    train_pred = clf.predict(X_train)
    train_acc = accuracy_score(y_train, train_pred)
    print('Train Accuracy of {:s} is {:.3f}', clf_name, train_acc)
    test_pred = clf.predict(X_test)
    test_acc = accuracy_score(y_test, test_pred)
    print('Test Accuracy of {:s} is {:.3f}', clf_name, test_acc)
```

```python
model = RandomForestClassifier(random_state=10)
from sklearn.model_selection import GridSearchCV

grid_vals = {'n_estimators': [13], 'max_depth': [22, 23, 24], 'max_features': ['auto', 'sqrt'], 'min_samples_split': [2, 5, 10]}

grid_rf = GridSearchCV(estimator=model, param_grid=grid_vals, scoring='accuracy', 
                       cv=10, refit=True, return_train_score=True) 
grid_rf.fit(X_train, y_train)

preds_train = grid_rf.best_estimator_.predict(X_train)
acc_train = accuracy_score(y_train, preds_train)
print("Train accuracy: ", acc_train)

preds_test = grid_rf.best_estimator_.predict(X_test)
acc_test = accuracy_score(y_test, preds_test)
print("Test accuracy: ", acc_test)
```
